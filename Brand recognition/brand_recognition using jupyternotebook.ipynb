```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt


cnn = tf.keras.models.load_model(r"C:\Users\ASUS\brand_recognition.h5")

#Test Image Visualization
import cv2
image_path =  r"C:\Users\ASUS\Desktop\Final Brand recognition\Brand_recogniton\test\Tide Washing Powder\images (4).jpeg"
# Reading an image in default mode
img = cv2.imread(image_path)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #Converting BGR to RGB
# Displaying the image
plt.imshow(img)
plt.title('Test Image')
plt.xticks([])
plt.yticks([])
plt.show()


image = tf.keras.preprocessing.image.load_img(image_path,target_size=(64,64)) #use same format that used for training
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr])  # Convert single image to a batch.
predictions = cnn.predict(input_arr)


print(predictions)


test_set = tf.keras.utils.image_dataset_from_directory(
    r"C:\Users\ASUS\Desktop\Final Brand recognition\Brand_recogniton\test",
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(64, 64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)



test_set.class_names


result_index = np.argmax(predictions) #Return index of max element
print(result_index)


# Displaying the image
plt.imshow(img)
plt.title('Test Image')
plt.xticks([])
plt.yticks([])
plt.show()


#Single image Prediction
print("It's a {}".format(test_set.class_names[result_index]))   



```
